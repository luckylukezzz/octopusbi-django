# **combining all of csv together**

```python
import pandas as pd
import glob

csv_directory = '/kaggle/input/ganison-dataset/*.csv'

csv_files = glob.glob(csv_directory)

dataframes = []

for file in csv_files:
    df = pd.read_csv(file) 
    dataframes.append(df)  

combined_df = pd.concat(dataframes, ignore_index=True)
```

```python
combined_df.shape
```
